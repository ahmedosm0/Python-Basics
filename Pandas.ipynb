# find players most runs against teams

```python
mask = data['batsman'] == "V Kohli"
player = data[mask]
runs = player.groupby('bowling_team')
TotalRuns = runs['batsman_runs']
# print(TotalRuns.size())
print(TotalRuns.sum().sort_values(ascending = False).head(3))
```

### find players most runs against teams
- using function

```python
def function(batsman_name):
    mask = data['batsman'] == batsman_name
    player = data[mask]
    runs = player.groupby('bowling_team')
    TotalRuns = runs['batsman_runs']
    # print(TotalRuns.size())
    print(TotalRuns.sum().sort_values(ascending = False).head(3))
```

```python
function("V Kohli")
```

# isin()
- find most destractive death over batsman in ipl
- strike rate = (num of runs/ num of balls)/100
- min batsman played 200 balls in over 16-20

```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("deliveries.csv")
```

```python
dataset.head(3)
```

```python
mask = dataset['over']>15
deathOver = dataset[mask]
deathOver
```

```python
batsman = deathOver.groupby('batsman')
allBatsmans = batsman.size()
mask = allBatsmans>200
requiredBatsmans = allBatsmans[mask]
requiredBatsmans
```

```python
batsmanList = requiredBatsmans.index.tolist()
batsmanList
```

```python
reqBatsmans = dataset[dataset['batsman'].isin(batsmanList)]
reqBatsmans
```

```python
reqBatsmans.groupby("batsman")["batsman_runs"].count()
```

```python
ballsPlayed = reqBatsmans.groupby("batsman").size()
ballsPlayed
```

```python
runsPlayed = reqBatsmans.groupby("batsman")['batsman_runs'].sum()
runsPlayed
```

```python
strikeRate = (runsPlayed / ballsPlayed)* 100
strikeRate.sort_values(ascending = False)
```

# merge()
- do anylasis to find top runner batsman for every season

```python
import numpy as np
import pandas as pd
```

```python
matchData = pd.read_csv("matches.csv")
deliveryData = pd.read_csv("deliveries.csv")
```

```python
matchData.head(2)
```

```python
deliveryData.head(2)
```

```python
newData = deliveryData.merge(matchData, left_on = "match_id", right_on = "id") # we merge both data sets. on basis of id.
newData.shape # both datasets columns are now added.
```

```python
deliveryData.shape
```

```python
matchData.shape
```

```python
newData.head()
```

```python
newData.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending = False).reset_index().drop_duplicates(subset = "season", keep= "first").sort_values('season')[['season','batsman']]
```

### step by step approach

```python
groupData = newData.groupby(['season','batsman'])
groupData
```

```python
new=groupData['batsman_runs'].sum().sort_values(ascending = False).reset_index() # reset index converts series to data frame
new
```

```python
reqData = new.drop_duplicates(subset = 'season',keep = 'first')
reqData
```

```python
final = reqData.sort_values('season')
final
```

# 1.filling missing values using dropna() method

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("deliveries.csv")
print(data.shape)
data.head()
```

```python
data1 = data.dropna() # removes the row if one missing value found row wise
print(data1.shape)
data1
```

```python
data2 = data.dropna(axis = 1) # removes the column, if one missing value found column wise.
print(data2.shape)
data2
```

```python
data2.dropna(axis=1, how='all').shape # how parameter removes columns, where all the values in column are null
```

```python
data3 = data.dropna(subset=['player_dismissed']) # it removes null values just from player_dismessed column
print(data3.shape)
data3
```

```python
data4 = data.dropna(subset = ['fielder'])
print(data4.shape)
data4
```

```python
data5 = data.dropna(subset=['player_dismissed','fielder'])
print(data5.shape)
data5
```
